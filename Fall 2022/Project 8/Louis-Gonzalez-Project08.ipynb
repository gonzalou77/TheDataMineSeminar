## Question 2

**Reasons Why Virtual Reasons are Necessary**
1) Version Control
    - Python packages are constantly being updated, however the scripts that use these packages are not updated to reflect package updates which creates version control issues. Virtual environments allow programmers and testers to utilize code that may have old versions of a package without rewriting the script to work with the current version of the package.

2) Work Around Privilege Installation Lockouts
    - There may be times in which users do not have administrative priveleges to install packages that may not be installed on the computer. A virtual environment allows a programmer to install packages within the virtual environment without the need or within the scope of the defined user priveleges which is a likely problem to encounter in the future.
